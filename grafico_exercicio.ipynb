### CriaÃ§Ã£o dos grÃ¡ficos

```python
fig, ax = plt.subplots(figsize=(10, 6))  # Set the figure size

# Calculate the counts of each state
counts = data['Name'].value_counts()

# Extract the state names (unique values)
name_values = counts.index

# Plotting the bar chart
ax.bar(name_values, counts)

ax.set_ylabel('Count')  # Label for y-axis
ax.set_xlabel('Name Air Quality')  # Label for x-axis
ax.set_title('Name')  # Title of the plot

plt.xticks(rotation=90)  # Rotate x-axis labels if needed for better readability

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))  # Set the figure size

# Calculate the counts of each state
counts = data['Time Period'].value_counts()

# Extract the state names (unique values)
time_period_values = counts.index

# Plotting the bar chart
ax.bar(time_period_values, counts)

ax.set_ylabel('Count')  # Label for y-axis
ax.set_xlabel('Time Period')  # Label for x-axis
ax.set_title('Time Period x Count')  # Title of the plot

plt.xticks(rotation=90)  # Rotate x-axis labels if needed for better readability

plt.show()
```
