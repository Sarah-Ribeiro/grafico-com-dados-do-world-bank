### Criação dos gráficos

```python
fig, ax = plt.subplots(figsize=(10, 6))  # Set the figure size

# Calculate the counts of each state
counts = data['Name'].value_counts()

# Extract the state names (unique values)
name_values = counts.index

# Plotting the bar chart
ax.bar(name_values, counts)

ax.set_ylabel('Count')  # Label for y-axis
ax.set_xlabel('Name Air Quality')  # Label for x-axis
ax.set_title('Name')  # Title of the plot

plt.xticks(rotation=90)  # Rotate x-axis labels if needed for better readability

plt.show()
```

É possível perceber que os níveis de dióxido de carbono são muito altos nos países onde a presença de partículas finas, produzidas pela combustão, é significativa. Essas partículas não são saudáveis, pois podem ser inaladas, prejudicando a saúde dos seres humanos.

```python
fig, ax = plt.subplots(figsize=(15, 6))  # Set the figure size

# Calculate the counts of each state
counts = data['Time Period'].value_counts()

# Extract the state names (unique values)
time_period_values = counts.index

# Plotting the bar chart
ax.bar(time_period_values, counts)

ax.set_ylabel('Count')  # Label for y-axis
ax.set_xlabel('Time Period')  # Label for x-axis
ax.set_title('Time Period x Count')  # Title of the plot

plt.xticks(rotation=90)  # Rotate x-axis labels if needed for better readability

plt.show()
```

Pelo gráfico acima é possível perceber que o período de tempo é maior entre os anos de 2007-2019, que foi o tempo em qua as industrias cresceram, em que a popualação mundial também teve um crescimento. Também é possível perceber que no verão é muito porpicio a qualidade do ar, que na maioria das vezes quando não chove, cria muita poluição pela se
